```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sistema internacional
```

```python
h = 6.626e-34      # constante de Planck [J·s]
c = 3.0e8          # velocidad de la luz en el vacío [m/s]
k = 1.381e-23      # constante de Boltzmann [J/K]
```

```python
# --- Ley de Planck ---
# lam : longitud de onda λ [m]
# T   : temperatura [K]
# devuelve I(λ) : intensidad espectral [W·m^-3·sr^-1]

def planck(lam, T):
    return (2*h*c**2)/(lam**5) * 1/(np.exp(h*c/(lam*k*T)) - 1)

# Constante de desplazamiento de Wien
b = 2.8977719e-3  # m·K

def lambda_max_wien(T, b=b):
    """Retorna la longitud de onda máxima (m) según la ley de Wien para una temperatura T."""
    return b / T

lam1 = np.linspace(1e-8,5e-6,1000)
lam2 = np.linspace(1e-8,5e-4,1000)

T1 = 6000
T2 = 10000
T3 = 100

# Longitudes de onda máximas (λ_max = b/T)
lambda_max1 = lambda_max_wien(T1)
lambda_max2 = lambda_max_wien(T2)
lambda_max3 = lambda_max_wien(T3)

print(f"λ_max (T={T1} K) = {lambda_max1*1e9:.2f} nm")
print(f"λ_max (T={T2} K) = {lambda_max2*1e9:.2f} nm")
print(f"λ_max (T={T3} K) = {lambda_max3*1e6:.2f} μm")

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(lam1*1e9, planck(lam1,T1), label="T = 6000 K")
plt.plot(lam1*1e9, planck(lam1,T2), label="T = 10000 K")
plt.axvline(lambda_max1*1e9, color="tab:blue", linestyle="--", alpha=0.7, label=f"λ_max T = {T1} K")
plt.axvline(lambda_max2*1e9, color="tab:orange", linestyle="--", alpha=0.7, label=f"λ_max T = {T2} K")
plt.xlabel("Longitud de onda λ [nm]")
plt.ylabel("Intensidad espectral I(λ) [W·m⁻³·sr⁻¹]")
plt.legend()

plt.subplot(1,2,2)
plt.plot(lam2*1e6, planck(lam2,T3), color="purple", label="T = 100 K")
plt.axvline(lambda_max3*1e6, color="purple", linestyle="--", alpha=0.7, label=f"λ_max T = {T3} K")
plt.xlabel("Longitud de onda λ [μm]")
plt.ylabel("Intensidad espectral I(λ) [W·m⁻³·sr⁻¹]")
plt.legend()

plt.tight_layout()
plt.show()
```
